```python
cbar_kws = {"orientation":"vertical", 
            "shrink":1,
            'extend':'max', 
            'extendfrac':0.2, 
            "ticks":np.arange(0,22), 
            "drawedges":True,
           }
#Drop specific record
data22 = data_n.drop(['American Samoa'])
"""plt.title("Who is responsible for global warming")
plt.xlabel("CO2 emission per year")
plt.ylabel("country")"""
plt.figure(figsize=(16,9))
ax = sns.heatmap(data22.head(20),vmin = 0, vmax = 21,annot=True,cbar_kws=cbar_kws)
ax.set(title="Who is responsible for global warming",
      xlabel="CO2 emission per year",
      ylabel="Country Name",)
```

```python
# sns heatmap correlation
 
plt.figure(figsize=(16,9))
 
ax = sns.heatmap(data22.corr(), annot = True,linewidths=1)
ax.tick_params(direction='out', length=9, width=1, colors='g',
               grid_color='b', grid_alpha=0.9,labelsize=10,labelcolor='red')
plt.title("Who is responsible for global warming")
plt.xlabel("CO2 emission per year")
plt.ylabel("country")
```
