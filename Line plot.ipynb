```python
plt.subplots(figsize=(10, 5))
sns.lineplot(data['size'],data['total_bill'],data = data , hue="sex",style = "sex",palette='Dark2',
             dashes = True, markers=[">","o"],legend = 'full')
plt.title("relationship b/w customer and bill" ,color = 'r', fontweight='bold',fontsize = "20")
plt.xlabel("size of the customers",color = 'g', fontweight='bold',fontsize = "15")
plt.ylabel("Total bill of the customers",color = 'b', fontweight='bold',fontsize = "15")
plt.show()
```
